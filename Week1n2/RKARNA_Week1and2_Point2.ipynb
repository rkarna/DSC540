```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("world-population.xlsm", engine = 'openpyxl')
```

```python
plt.figure(figsize =(16,8))
x_axis = list(data.Year.values.tolist())
y_axis = list(data.Population.values.tolist())
plt.plot(x_axis,y_axis, color = 'b')
plt.xlabel("YEAR")
plt.ylabel("POPULATION")
plt.title("LINE PLOT - Population VS Year")
plt.show()
```
